```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,RobustScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

```python
df=pd.read_csv('weatherAUS.csv')
```

```python
df['RainToday'].unique()
```

```python
df.head(n=5)
```

```python
df.info()
```

```python
df=df.dropna(subset='RainToday')
```

```python
df=df.drop(columns='RainTomorrow')
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Date']=df['Date'].astype(int)
```

```python
df.head()
```

```python
df['Evaporation']=df['Evaporation'].fillna(df['Evaporation'].median())
df['Rainfall']=df['Rainfall'].fillna(df['Rainfall'].median())
df['MinTemp']=df['MinTemp'].fillna(df['MinTemp'].mean())
df['MaxTemp']=df['MaxTemp'].fillna(df['MaxTemp'].mean())
df['Sunshine']=df['Sunshine'].fillna(df['Sunshine'].mean())
df['WindGustDir']=df['WindGustDir'].fillna(df['WindGustDir'].mode()[0])
df['WindGustSpeed']=df['WindGustSpeed'].fillna(df['WindGustSpeed'].median())
df['WindDir9am']=df['WindDir9am'].fillna(df['WindDir9am'].mode()[0])
df['WindDir3pm']=df['WindDir3pm'].fillna(df['WindDir3pm'].mode()[0])
df['WindSpeed9am']=df['WindSpeed9am'].fillna(df['WindSpeed9am'].median())
df['WindSpeed3pm']=df['WindSpeed3pm'].fillna(df['WindSpeed3pm'].median())
df['Humidity9am']=df['Humidity9am'].fillna(df['Humidity9am'].median())
df['Humidity3pm']=df['Humidity3pm'].fillna(df['Humidity3pm'].median())
df['Pressure9am']=df['Pressure9am'].fillna(df['Pressure9am'].median())
df['Pressure3pm']=df['Pressure3pm'].fillna(df['Pressure3pm'].median())
df['Cloud9am']=df['Cloud9am'].fillna(df['Cloud9am'].median())
df['Cloud3pm']=df['Cloud3pm'].fillna(df['Cloud3pm'].median())
df['Temp9am']=df['Temp9am'].fillna(df['Temp9am'].median())
df['Temp3pm']=df['Temp3pm'].fillna(df['Temp3pm'].median())
```

```python
df['Evaporation'].unique().sum()
```

```python
df['RainToday'].unique()
```

```python
df['RainToday']=df['RainToday'].map({'No':0, 'Yes':1})
```

```python
encoder=OneHotEncoder(sparse_output=False)
values=encoder.fit_transform(df[['Location','WindGustDir','WindDir9am','WindDir3pm']])
```

```python
names = encoder.get_feature_names_out()
```

```python
en_df = pd.DataFrame(values, columns=names,index=df.index)
df = pd.concat([df, en_df], axis=1)
df = df.drop(columns=['Location','WindGustDir','WindDir9am','WindDir3pm'])
df
```

```python
x=df.drop(columns='RainToday')
y=df.RainToday
```

```python
num_cols=x.select_dtypes(include='number').columns
cat_cols=x.select_dtypes(include='object').columns
```

```python
over=SMOTE(sampling_strategy=0.9)
x_sam, y_sam=over.fit_resample(x,y)
y_sam.value_counts()
```

```python
under=RandomUnderSampler()
x,y=under.fit_resample(x_sam,y_sam)
y.value_counts()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
main_pipeline=Pipeline(
    steps=[
        ('scaler',RobustScaler()),
        ('model',KNeighborsClassifier())
    ]
)

main_pipeline.fit(xtrain,ytrain)
```

```python
y_pred=main_pipeline.predict(xtrain)
print(classification_report(ytrain,y_pred))
```

```python
y_test=main_pipeline.predict(xtest)
print(classification_report(ytest,y_test))
```

```python
main_pipeline=Pipeline(
    steps=[
        ('scaler',RobustScaler()),
        ('model',XGBClassifier())
    ]
)

main_pipeline.fit(xtrain,ytrain)
```

```python
y_pred=main_pipeline.predict(xtrain)
print(classification_report(ytrain,y_pred))
```

```python
y_test=main_pipeline.predict(xtest)
print(classification_report(ytest,y_test))
```

```python
main_pipeline=Pipeline(
    steps=[
        ('scaler',RobustScaler()),
        ('model',LogisticRegression(max_iter=500))
    ]
)

main_pipeline.fit(xtrain,ytrain)
```

```python
y_pred=main_pipeline.predict(xtrain)
print(classification_report(ytrain,y_pred))
```

```python
y_test=main_pipeline.predict(xtest)
print(classification_report(ytest,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier 
main_pipeline=Pipeline(
    steps=[
        ('scaler',RobustScaler()),
        ('model',DecisionTreeClassifier())
    ]
)

main_pipeline.fit(xtrain,ytrain)
```

```python
y_pred=main_pipeline.predict(xtrain)
print(classification_report(ytrain,y_pred))
```

```python
y_test=main_pipeline.predict(xtest)
print(classification_report(ytest,y_test))
```
